```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
pdd = pd.read_excel('ki.xls')
pdd
```

```python
ddff = pd.read_excel('ki_udp.xls')
ddff
```

```python
ax = plt.plot(pdd['Measurement (meters)'],pdd['S/N ratio MV'], 'o--', label='TCP')
ax = plt.plot(ddff['Measurement (Meters)'],ddff['S/N ratio MV'], 'D-', label='UDP')
plt.title('SNR vs Distance')
plt.legend()
plt.grid()
plt.xlabel('Distance(m)')
plt.ylabel('Signal-Noise ratio [db]')
plt.show()

```

```python
ax1 = plt.plot(pdd['Measurement (meters)'],pdd['Throughput MV'], 'o--', label='TCP')
ax1 = plt.plot(ddff['Measurement (Meters)'],ddff['Throughput MV'], 'D-', label='UDP')
plt.title('Throughput vs Distance')
plt.legend()
plt.grid()
plt.xlabel('Distance(m)')
plt.ylabel('Throughput (Mbit/s)')
plt.show()
```

```python
ax2 = plt.plot(pdd['Sorted SNR'],pdd['Throughput MV'], 'o--', label='TCP')
ax2 = plt.plot(ddff['Sorted SNR'],ddff['Throughput MV'], 'D-', label='UDP')
plt.title('Throughput vs SNR(Sorted)')
plt.legend()
plt.grid()
plt.xlabel('Signal-Noise ratio [db]')
plt.ylabel('Throughput (Mbit/s)')
plt.show()
```

```python
ax3 = plt.plot(pdd['Measurement (meters)'],pdd['Latency Time'], 'o--', label='TCP')
ax3 = plt.plot(ddff['Measurement (Meters)'],ddff['Latency Time'], 'D-', label='UDP')
plt.title('Latency time vs Distance')
plt.legend()
plt.grid()
plt.xlabel('Distance(m)')
plt.ylabel('Latency Time')
plt.show()
```
