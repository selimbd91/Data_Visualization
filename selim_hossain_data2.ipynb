```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

x = np.random.normal(loc= 3.0, scale= 10, size= 100)

ax = sns.distplot(x)
ax.set_xlabel('Random Variable')
ax.set_ylabel('Probability Density Function')
ax.set_title('Normal Distributation Function')

plt.show()
```
